```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
df=pd.read_csv('C:/Users/zahie/OneDrive/מסמכים/GitHub/OkCupid/profiles.csv')
```

Income per age

```python
sns.lmplot(x='income',y='age', data=df[(df['income']>0) & (df['income']<400000)],order=2)
```

Activity-hour per age

```python
df['hour']=df['last_online'].apply(lambda x: int(x[11:13]))
df['roundage']=df['age'].apply(lambda x: np.round(x))
sns.factorplot(y='age',x='hour',data=df)
```

Job per age

```python
s=df[df['income']>0].pivot_table(index='job',columns='age', values='income', aggfunc='count')
sns.heatmap(s)
```

Income per etnicity

```python
df['race']=df['ethnicity'].apply(lambda x: str(x).split(',')[0])
sns.barplot(x='income',y='race',data=df[df['income']>0])
```

status per age

```python
sns.barplot(y='status',x='age',hue='sex',data=df)
```
