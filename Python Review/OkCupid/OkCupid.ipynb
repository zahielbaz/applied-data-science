```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
df=pd.read_csv('C:/Users/zahie/OneDrive/מסמכים/GitHub/OkCupid/profiles.csv')
```

First, see how religion distributed

```python
df['religion1']=df['religion'].apply(lambda x: str(x))
df['religion1']=df['religion1'].apply(lambda x: x.split()[0])
df['jew']=df['religion1'].apply(lambda x: 'jew' if x=='judaism' else 'other')
byJews=df.groupby('jew')
byReligion=df.groupby('religion1')
sns.countplot(x='religion1',data=df)
#c=pd.DataFrame(byReligion['age'].count())
#np.array(c.values)
#plt.plot(np.array(c.values))
```

Jews are half male and half female

```python
a=byJews['sex'].value_counts(normalize=True,sort=False)
a.unstack().plot(kind='bar')
```

Jews around the age of 60 looking for a partner

```python
b=byJews['age'].value_counts(normalize=True,sort=False)
#plt.plot(x=np.float32(a['jew'].axes)[0],y=np.float32(a['jew'].values))
x = np.float32(b['jew'].axes)[0]
y = np.float32(b['jew'].values)
x1 = np.float32(b['other'].axes)[0]
y1 = np.float32(b['other'].values)
plt.plot(x, y)
plt.plot(x1, y1)
plt.show()
```

Males are higher.
